```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dx/dt
def model(x,t,k):
    dxdt = -k * x
    return dxdt

# initial condition
x0 = 5

# time points
t = np.linspace(0,20)

# solve ODEs
k = 0.1
x1 = odeint(model,x0,t,args=(k,))
k = 0.2
x2 = odeint(model,x0,t,args=(k,))
k = 0.5
x3 = odeint(model,x0,t,args=(k,))

# plot results
plt.plot(t,x1,'r-',linewidth=2,label='k=0.1')
plt.plot(t,x2,'b--',linewidth=2,label='k=0.2')
plt.plot(t,x3,'g:',linewidth=2,label='k=0.5')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dx/dt
def model(x,t,k):
    dxdt = -k * x
    return dxdt

# initial condition
x0 = 5

# time points
t = np.linspace(0,20)

# solve ODE
k = 0.1
x1 = odeint(model,x0,t,args=(k,))

k = 0.2
x2 = odeint(model,x0,t,args=(k,))

k = 0.5
x3 = odeint(model,x0,t,args=(k,))

# plot results
plt.plot(t,x1)
plt.plot(t,x2)
plt.plot(t,x3)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.show()
```
