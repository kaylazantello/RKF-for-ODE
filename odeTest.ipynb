```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def model(s,n):
    f = 0.8
    #dsdn = (f / n**2) / (((1 - f) + (f / n))**2)
    dsdn = s**2 * f / n**2
    return dsdn

s0 = 1

n = np.linspace(1,10)

s = odeint(model, s0, n)

plt.plot(n,s)
plt.xlabel('parallel processors')
plt.ylabel('s(n)')
plt.show()
```

```python
def model(s,n,f):
    dsdn = (f / n**2) / (((1 - f) + (f / n))**2)
    return dsdn

s0 = 1

n = np.linspace(1,10)

f = 0.25
s1 = odeint(model, s0, n, args=(f,))
f = 0.45
s2 = odeint(model, s0, n, args=(f,))
f = 0.8
s3 = odeint(model, s0, n, args=(f,))

plt.plot(n,s1,label='f = 0.25')
plt.plot(n,s2,label='f = 0.45')
plt.plot(n,s3,label='f = 0.8')
plt.xlabel('# of parallel processors')
plt.ylabel('speedup')
plt.legend()
plt.show()
```

```python
# function that returns dx/dt
def model(x,t):
    k = 1
    dxdt = k * x
    return dxdt

# initial condition
x0 = 1

# time points
t = np.linspace(0,20)

# solve ODE
x = odeint(model,x0,t)

# plot results
plt.plot(t,x)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.show()
```
